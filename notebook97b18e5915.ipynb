```python
import requests
from bs4 import BeautifulSoup

# This is the URL where the Google Docs document is published
url = "https://docs.google.com/document/d/e/2PACX-1vSHesOf9hv2sPOntssYrEdubmMQm8lwjfwv6NPjjmIRYs_FOYXtqrYgjh85jBUebK9swPXh_a5TJ5Kl/pub"

# Get the content from the URL
response = requests.get(url)

# Check if the request was successful
response.raise_for_status()

# Use BeautifulSoup to read and parse the content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table in the parsed content
table = soup.find('table')

# Create an empty list to hold the data
data = []

# Initialize variables to keep track of the maximum x and y values
max_x, max_y = 0, 0


# Extract data from the table
# Loop through each row of the table, skipping the first one
for row in table.find_all('tr')[1:]:
    # Get all cells in the row and remove any extra spaces around the text
    cells = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
    
    # Extract and convert the cell values to the appropriate types
    x = int(cells[0])  # Convert the first cell to an integer
    char = cells[1]    # The second cell is a string
    y = int(cells[2])  # Convert the third cell to an integer
    
    # Add the values as a tuple to the data list
    data.append((x, char, y))
    
    # Update the maximum values for x and y
    if x > max_x:
        max_x = x
    if y > max_y:
        max_y = y


# Create a 2D list (array) with empty spaces
array_2d = []

# Loop through the range from 0 to max_y (inclusive) to create rows
for _ in range(max_y + 1):
    # For each row, create a list of empty spaces
    row = [' ' for _ in range(max_x + 1)]
    # Add the row to the 2D list
    array_2d.append(row)


# Fill the 2D array with characters from the data
for x, char, y in data:
    # Place the character in the correct position in the 2D array
    array_2d[y][x] = char

# Print each row of the 2D array
for row in array_2d:
    # Join all elements in the row into a single string and print it
    print(''.join(row))

```
